```python
import numpy as np
import tellurium as te
import matplotlib.pyplot as plt
```

```python
antstr = '''
# CSF import/decay
-> CSF; k_imp * CSFex * Opp
CSF ->; k_dCSF * CSF

# ComX binding
Pb: ComPb -> ComPf + ComX; k_fP * ComPb
Pf: ComPf + ComX -> ComPb; k_bP * ComPf * ComX

# Rap B
B1: CSF + RapB -> RapBx; k_B * CSF * RapB
B2: RapBx -> CSF + RapB; k_dB * RapBx

# RapC
B1: CSF + RapC -> RapCx; k_C * CSF * RapC
B2: RapCx -> CSF + RapC; k_dC * RapCx

# Spo0F
pSpo: Spo0F -> pSpo0F; k_F * Spo0F * (KinA^h / (KK^h + KinA^h))
Spo: pSpo0F -> Spo0F; k_dF * pSpo0F * (RapBx^h / (KB^h + RapBx^h))

# ComA
A1: ComA -> pComA; k_A * ComA * (ComPb^h / (KP^h + ComPb^h))
A2: pComA -> ComA; k_dA * pComA * (PhoX^h / (KX^h + PhoX^h))

# Competence response is tempered by RapC
res: -> Compe; k_comp * pComA * (KinA^h / (KK^h + KinA^h)) * (KC ^h / (KC^h + RapCx^h))


# Initial concentrations
# Fixed species
$CSFex = 0.01
$Opp = 0.5
$ComX = 0.01
$KinA = 0.1
$PhoX = 0.1

# Changing Species
# signals
CSF = 0
ComPf = 0.5
ComPb = 0

# mediators
RapB = 0.8
RapC = 0.8
RapBx = 0
RapCx = 0

# responses
Spo0F = 0.25
pSpo0F = 0.25

ComA = 0.25
pComA = 0.25

Compe = 0

# Rate Constants
# just to have a default value
k = 1.0

# CSF
k_imp = k
k_dCSF = k

# ComP
k_fP = k
k_bP = k

# RapB
k_B = k
k_dB = k

# RapC
k_C = k
k_dC = k

# Spo0F
k_F = k
k_dF = k

# ComA
k_A = k
k_dA = k

# Competence
k_comp = k

# Michaelis Constants
KK = 0.5
KB = 0.2
KP = 0.5
KX = 0.5
KC = 0.6
KA = 0.6

# Hill
h = 4
'''

mod = te.loada(antstr)
```
